# Data Pre-Processing

```python
encoded_train = train_data.copy()
encoded_test_X = test_X.copy()
encoded_test_y = test_y.copy()
```

```python
def delete_col(df,col):
    for i in col:
        df.drop(i,axis = 1,inplace = True)
```

```python
col = ['Id','Alley','Street','MasVnrType','MiscFeature','Fence','PoolQC','FireplaceQu','MoSold']
delete_col(encoded_train,col)
delete_col(encoded_test_X,col)
```

```python
col = ['Id']
delete_col(encoded_test_y,col)
```

```python
def fill_null(df,col):
    for i in col:
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)
        iqr = q3 - q1
        lower_lim = q1 - 1.5*iqr
        upper_lim = q3 + 1.5*iqr
        without_outlier = df[i].apply(lambda x: None if x < lower_lim or x > upper_lim else x)
        without_outlier.fillna(without_outlier.mean(),inplace = True)
        df[i] = without_outlier
def fill_obj(df,col):
    for i in col:
        df[i].fillna(df[i].mode()[0],inplace = True)
```

```python
def int_col(df):
    col = []
    for i in df.columns:
        if df[i].dtype != 'O':
            col.append(i)
    return col
def obj_col(df):
    col = []
    for i in df.columns:
        if df[i].dtype == 'O':
            col.append(i)
    return col
```

```python
col = int_col(encoded_train)
fill_null(encoded_train,col)
```

```python
col = int_col(encoded_test_X)
fill_null(encoded_test_X,col)
```

```python
col = obj_col(encoded_train)
fill_obj(encoded_train,col)
```

```python
col = obj_col(encoded_test_X)
fill_obj(encoded_test_X,col)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def obj_to_int(df,col):
    for i in col:
        label = LabelEncoder()
        df[i] = label.fit_transform(df[i])
```
